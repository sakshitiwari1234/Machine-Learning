### Support Vector Machine (SVM)



SVM is a classification problem to classify things

in this lecture we shall classify , and know the features of an iris flower having sepal , petal length and width repsectively.


when we draw a spearation boundary to classify the graph , all may occur same , then we can decide which to decdie on the basis of margiin between the diffeerent , boundaries.

so, whenever , there we perform classification problem on a graph with two distinct features and draw multiple lines separating the line grpahs , we need to choose that which graph line fits the best , and divide equally, in this case we decide with the help of suppport vector plots. the line having vecotr(plot) which has maximum margin from the line is considered to be the best ,  fit line to draw and divide the classification.

1. IN A TWO DIMENSIONAL PLANE HAVING , ONLY TWO FEATURES TO DISTINGUISH ,  -> THERE IS A LINE DRAWN,

2. IN A TWO DIMENSIONAL PLANE , FEATURES ARE DISTNGUISH BY A PLANE.

3. TO DISTINGUISH, MULTIPLE, N DIMENSIONS , MATHEMATICAL HYPERBOLA IS USED.

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame( iris.data, columns=iris.feature_names)
```

```python
df
```

```python
df['target'] = iris.target
df.head()
```

```python
iris.target_names
```

```python
df[df.target==1].head()
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target == 2 ]
```

```python
plt.xlabel('petal length in cm')
plt.ylabel('petal width in cm')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color = 'green', marker='+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color= 'blue', marker= '+')
```
