# Decision Tree classification works on both of the data in categorical and numerical data as well.

```python
import pandas as pd
df = pd.read_csv("/content/decision_tree.csv")
df
```

```python
df.head()
```

```python
inputs = df.drop('salary', axis='columns')
target =df['salary']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs ['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n']  = le_company.fit_transform(inputs['job'])
inputs['degree_n'] = le_company.fit_transform(inputs['degree'])
inputs.head()
```

```python
input_n = inputs.drop(['company', 'degree', 'job' ], axis = 'columns')
input_n
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(input_n, target)
```

```python
model.score(input_n, target)
```

```python
model.predict([[2,2,1]])
```
